```python
df2 = pd.read_csv('/content/drive/MyDrive/unsw_data/UNSW-NB15_1.csv', header=None)
```

```python
print(df2.head())
```

```python
df2 = df2.iloc[:, [0,1,2,3,4,6,7,8,48]]
df2.columns = ['srcip','sport','dstip','dsport','proto','dur','sbytes','dbytes','label']
```

```python
df2['label'] = df2['label'].apply(lambda x: 'attack' if x == 1 else 'normal')
```

```python
def to_text2(row):
    return f"""
    Source IP: {row['srcip']},
    Destination IP: {row['dstip']},
    Protocol: {row['proto']},
    Duration: {row['dur']},
    Source bytes: {row['sbytes']},
    Destination bytes: {row['dbytes']}
    """

df2['text'] = df2.apply(to_text2, axis=1)
```

```python
df_test2 = df2.sample(20, random_state=42)
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

def classify_gpt2(text):
    prompt = f"""You are a cybersecurity expert.
Classify the following network traffic as Normal or Attack.

{text}

Answer:"""

    result = generator(
        prompt,
        max_new_tokens=3,
        do_sample=False
    )

    output = result[0]['generated_text'].lower()

    if "normal" in output:
        return "normal"
    elif "attack" in output:
        return "attack"
    else:
        return "normal"
```

```python
df_test2['prediction'] = df_test2['text'].apply(classify_gpt2)
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report

print("Accuracy:", accuracy_score(df_test2['label'], df_test2['prediction']))
print("F1:", f1_score(df_test2['label'], df_test2['prediction'], pos_label='attack'))

print(classification_report(df_test2['label'], df_test2['prediction']))
```

```python
accuracies = []
f1_scores = []

for seed in [1, 7, 42, 100, 200]:
    df_test2 = df2.sample(20, random_state=seed)
    df_test2['prediction'] = df_test2['text'].apply(classify_gpt2)

    acc = accuracy_score(df_test2['label'], df_test2['prediction'])
    f1 = f1_score(df_test2['label'], df_test2['prediction'], pos_label='attack')

    print(f"Seed {seed} -> Accuracy: {acc}, F1: {f1}")

    accuracies.append(acc)
    f1_scores.append(f1)
```

```python
import numpy as np

print("Average Accuracy:", np.mean(accuracies))
print("Average F1:", np.mean(f1_scores))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(df_test2['label'], df_test2['prediction'])

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['normal','attack'],
            yticklabels=['normal','attack'])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - GPT (UNSW)")

plt.savefig('/content/drive/MyDrive/confusion_matrix_gpt_unsw.png')

plt.show()
```

```python
import matplotlib.pyplot as plt

seeds = [1,7,42,100,200]
accuracies = accuracies
f1_scores = f1_scores

plt.figure(figsize=(8,5))

plt.plot(seeds, accuracies, marker='o', label='Accuracy')
plt.plot(seeds, f1_scores, marker='s', label='F1 Score')

plt.title("Performance Across Samples - GPT")
plt.xlabel("Sample (Seed)")
plt.ylabel("Score")
plt.legend()

plt.savefig('/content/drive/MyDrive/multi_sample_results.png')

plt.show()
```
